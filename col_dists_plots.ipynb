## Make plots for poster

```python
# imports
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

print('Done imports!')
```

```python
# create column header names
inp_names = list(range(60))
names = inp_names + ['Rock_or_Mine']

# read in data
df = pd.read_csv('sonar.all-data', names=names)

# change target to binary (Mine = 1)
df['Rock_or_Mine'] = pd.get_dummies(df.Rock_or_Mine)['M']

# show
df
```

```python
# separate rock and mine data
df_rock = df[df.Rock_or_Mine == 0]
df_mine = df[df.Rock_or_Mine == 1]
```

```python
# instantiate figure
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

# add column 0 histogram
ax[0][0].hist(df_rock[0], color='red', alpha=0.5, label='rock', density=True)
ax[0][0].hist(df_mine[0], color='blue', alpha=0.3, label='mine', density=True)
ax[0][0].set_xlabel('Sonar Data from Angle 1')
ax[0][0].set_ylabel('Density')
ax[0][0].legend()

# add column 19 histogram
ax[1][0].hist(df_rock[19], color='red', alpha=0.5, label='rock', density=True)
ax[1][0].hist(df_mine[19], color='blue', alpha=0.3, label='mine', density=True)
ax[1][0].set_xlabel('Sonar Data from Angle 20')
ax[1][0].set_ylabel('Density')
ax[1][0].legend()

# add column 39 histogram
ax[0][1].hist(df_rock[39], color='red', alpha=0.5, label='rock', density=True)
ax[0][1].hist(df_mine[39], color='blue', alpha=0.3, label='mine', density=True)
ax[0][1].set_xlabel('Sonar Data from Angle 40')
ax[0][1].set_ylabel('Density')
ax[0][1].legend()

# add column 59 histogram
ax[1][1].hist(df_rock[59], color='red', alpha=0.5, label='rock', density=True)
ax[1][1].hist(df_mine[59], color='blue', alpha=0.3, label='mine', density=True)
ax[1][1].set_xlabel('Sonar Data from Angle 60')
ax[1][1].set_ylabel('Density')
ax[1][1].legend()

# set title
fig.suptitle('Column Distributions for Sonar Data')

# make nice layout
plt.tight_layout()

# save figure
plt.savefig('col_dists.png')

# show
plt.show()
```
